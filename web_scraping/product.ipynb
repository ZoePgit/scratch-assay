```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.dienmayxanh.com/bep-tu-don#c=1982&p=265815&o=2&pi=0'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
product_title = soup.find('div', class_= 'product-title')
```

```python
print(product_title)
```
